```python
import cv2
import pandas as pd
from datetime import datetime
from ultralytics import YOLO
import time
import os

# 1. Load the model
model = YOLO(r'C:\Users\LOQ\runs\detect\train5\weights\best.pt')

# Custom Mapping to your folder names
custom_names = {
    'motorcycle': 'Bike', 'bicycle': 'Bike', 'airplane': 'plane',
    'bus': 'bus', 'car': 'Car', 'boat': 'boat', 'train': 'train', 'truck': 'bus'
}

# Only these names will be counted in the final tally
vehicle_classes = ['Bike', 'boat', 'bus', 'Car', 'plane', 'tank', 'train']

cap = cv2.VideoCapture(0)
detection_logs = []
last_logged_time = {}

# File numbering logic
file_index = 1
while os.path.exists(f"Detection_Report_{file_index}.xlsx"):
    file_index += 1

print(f"System ON. Run ID: {file_index}. Tracking Vehicles...")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break

    results = model(frame, stream=True, conf=0.4)
    current_now = datetime.now()

    for r in results:
        for box in r.boxes:
            orig_label = model.names[int(box.cls[0])]
            display_name = custom_names.get(orig_label, orig_label)
            
            # Draw boxes
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, display_name, (x1, y1 - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

            # Log once every 2 seconds
            if display_name not in last_logged_time or (time.time() - last_logged_time[display_name] > 2):
                detection_logs.append({
                    "Date": current_now.strftime("%Y-%m-%d"),
                    "Time": current_now.strftime("%H:%M:%S"),
                    "Object Type": display_name 
                })
                last_logged_time[display_name] = time.time()

    cv2.imshow("Real-Time Analytics System", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): break

cap.release()
cv2.destroyAllWindows()

# --- Saving with Summary Table ---
if detection_logs:
    df = pd.DataFrame(detection_logs)
    filename = f"Detection_Report_{file_index}.xlsx"
    
    # Create the Vehicle Summary Count
    # We only count items that are in our vehicle_classes list
    summary_data = df[df['Object Type'].isin(vehicle_classes)]['Object Type'].value_counts().reset_index()
    summary_data.columns = ['Vehicle Type', 'Total Count']

    # Use ExcelWriter to save both the log and the summary in the same sheet
    with pd.ExcelWriter(filename, engine='openpyxl') as writer:
        df.to_excel(writer, index=False, sheet_name='Detection Log')
        
        # We write the summary starting a few rows below the main data
        summary_start_row = len(df) + 3
        summary_data.to_excel(writer, index=False, sheet_name='Detection Log', startrow=summary_start_row)
        
        # Add a title for the summary
        sheet = writer.sheets['Detection Log']
        sheet.cell(row=summary_start_row, column=1).value = "VEHICLE SUMMARY TALLY"

    print(f"\nSUCCESS: Report {filename} saved with Vehicle Tally.")
    print(summary_data.to_string(index=False)) # Shows counts in your console too
else:
    print("\nNo objects detected.")
```
